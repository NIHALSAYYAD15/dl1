### 5. Implement feed forward single layer perceptron with suitable example.


```python
def and_gate_perceptron(x1, x2):
    # Set weights and bias
    w1, w2, b = 0.5, 0.5, -0.7
    
    # Calculate the weighted sum
    weighted_sum = w1 * x1 + w2 * x2 + b
    
    # Apply the step function
    output = 1 if weighted_sum >= 0 else 0
    
    return output

# Test the AND gate perceptron
print("[0,0] > " + str(and_gate_perceptron(0, 0)))  # Output: 0
print("[0,1] > " + str(and_gate_perceptron(0, 1)))  # Output: 0
print("[1,0] > " + str(and_gate_perceptron(1, 0)))  # Output: 0
print("[1,1] > " + str(and_gate_perceptron(1, 1)))  # Output: 1
```
